```python
credits = pd.read_csv('/Users/shreyarai/Documents/Data analysis project/Recommendation system/archive/tmdb_5000_credits.csv')
movies = pd.read_csv('/Users/shreyarai/Documents/Data analysis project/Recommendation system/archive/tmdb_5000_movies.csv')
```

```python
# TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = 'english')
movies['overview'] = movies['overview'].fillna("")

#Vector Space Model
tfidf_matrix = tfidf.fit_transform(movies ['overview'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#indices = pd.Series(movies.index, index = movies['original_title']).drop_dublicates()
#indices
```

```python
indices = pd.Series(movies.index, index=movies['original_title']).drop_duplicates()
indices
```

```python
indices['The Dark Knight Rises']
```

```python
def get_recommendations(title, cosine_sim = cosine_sim):
    idx = indices[title]
    sim_scores = enumerate(cosine_sim[idx])
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    sim_scores = sim_scores[1:11]
    sim_index =[i[0]for i in sim_scores]
    print(movies['original_title'].iloc[sim_index]) 
```

```python
get_recommendations('The Dark Knight Rises')
```
